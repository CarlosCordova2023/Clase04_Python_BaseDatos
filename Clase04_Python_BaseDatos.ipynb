```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Datos simulados
crecimiento_economico = np.array([2.5, 3.1, 4.0, 4.5, 5.0, 5.5]).reshape(-1, 1)
ingresos_fiscales = np.array([5000, 5200, 5800, 6000, 6300, 6700])

# Crear y entrenar el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(crecimiento_economico, ingresos_fiscales)

# Predecir ingresos futuros
ingresos_predichos = modelo.predict(crecimiento_economico)

# Visualización de los resultados
plt.scatter(crecimiento_economico, ingresos_fiscales, color='blue', label='Datos Reales')
plt.plot(crecimiento_economico, ingresos_predichos, color='red', label='Modelo de Regresión')
plt.title('Regresión Lineal: Crecimiento Económico vs Ingresos Fiscales')
plt.xlabel('Crecimiento Económico (%)')
plt.ylabel('Ingresos Fiscales (millones de CLP)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Datos simulados
retorno_proyecto = np.array([10, 5, 15, 3, 12, 6, 8, 4]).reshape(-1, 1)  # En porcentaje
costo_proyecto = np.array([100, 50, 150, 30, 120, 60, 80, 40])  # En millones de CLP
resultados = np.array([1, 0, 1, 0, 1, 0, 1, 0])  # 1: Rentable, 0: No Rentable

# Combinar las variables
datos = np.hstack((retorno_proyecto, costo_proyecto.reshape(-1, 1)))

# Dividir los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(datos, resultados, test_size=0.3, random_state=42)

# Crear y entrenar el modelo de árbol de decisión
modelo_arbol = DecisionTreeClassifier()
modelo_arbol.fit(X_train, y_train)

# Predecir resultados en los datos de prueba
predicciones = modelo_arbol.predict(X_test)

# Evaluar el modelo
print(f"Precisión del modelo: {accuracy_score(y_test, predicciones) * 100:.2f}%")
```

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# Datos simulados: Gasto y número de empleados por departamento
gasto = [1000, 1500, 2000, 2500, 3000]
empleados = [50, 100, 150, 200, 250]

# Combinar los datos
datos = np.array(list(zip(gasto, empleados)))

# Aplicar K-Means para agrupar los departamentos
kmeans = KMeans(n_clusters=2)
kmeans.fit(datos)

# Resultados del clustering
etiquetas = kmeans.labels_

# Visualización de los clusters
plt.scatter(gasto, empleados, c=etiquetas, cmap='viridis')
plt.title('Clustering de Departamentos por Gasto y Número de Empleados')
plt.xlabel('Gasto (millones de CLP)')
plt.ylabel('Número de Empleados')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos simulados de gastos e ingresos fiscales
data = {
    'Año': [2015, 2016, 2017, 2018, 2019, 2020],
    'Gastos': [50, 55, 60, 65, 70, 75],  # En millones de CLP
    'Ingresos': [45, 50, 55, 60, 65, 70],  # En millones de CLP
}

df = pd.DataFrame(data)

# Preparar los datos
X = df[['Gastos']]
y = df['Ingresos']

# Crear y ajustar el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X, y)

# Predicciones
predicciones = modelo.predict(X)

# Visualización
plt.scatter(df['Gastos'], df['Ingresos'], color='blue', label='Datos Reales')
plt.plot(df['Gastos'], predicciones, color='red', label='Modelo de Regresión')
plt.title('Relación entre Gastos e Ingresos Fiscales')
plt.xlabel('Gastos (millones de CLP)')
plt.ylabel('Ingresos (millones de CLP)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos simulados de gastos e ingresos fiscales
data = {
    'Año': [2015, 2016, 2017, 2018, 2019, 2020],
    'Gastos': [50, 55, 60, 65, 70, 75],  # En millones de CLP
    'Ingresos': [45, 50, 55, 60, 65, 70],  # En millones de CLP
}

df = pd.DataFrame(data)

# Preparar los datos
X = df[['Gastos']]
y = df['Ingresos']

# Crear y ajustar el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X, y)

# Predicciones
predicciones = modelo.predict(X)

# Visualización
plt.scatter(df['Gastos'], df['Ingresos'], color='blue', label='Datos Reales')
plt.plot(df['Gastos'], predicciones, color='red', label='Modelo de Regresión')

# Agregar valores a los puntos
for i in range(len(df)):
    plt.text(df['Gastos'][i], df['Ingresos'][i], f'({df["Gastos"][i]}, {df["Ingresos"][i]})', fontsize=9, ha='right')

plt.title('Relación entre Gastos e Ingresos Fiscales')
plt.xlabel('Gastos (millones de CLP)')
plt.ylabel('Ingresos (millones de CLP)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Datos simulados de proyectos
data = {
    'Tasa_Retorno': [10, 5, 15, 3, 12, 6, 8, 4],  # En %
    'Riesgo': [5, 7, 3, 9, 4, 8, 6, 7],  # Índice de riesgo
    'Rentable': [1, 0, 1, 0, 1, 0, 1, 0],  # 1: Rentable, 0: No Rentable
}

df = pd.DataFrame(data)

# Separar características y etiquetas
X = df[['Tasa_Retorno', 'Riesgo']]
y = df['Rentable']

# Dividir datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear y ajustar el modelo de árbol de decisión
modelo_arbol = DecisionTreeClassifier()
modelo_arbol.fit(X_train, y_train)

# Predicciones
predicciones = modelo_arbol.predict(X_test)

# Evaluar el modelo
precisión = accuracy_score(y_test, predicciones)
print(f"Precisión del modelo de clasificación: {precisión * 100:.2f}%")
```
